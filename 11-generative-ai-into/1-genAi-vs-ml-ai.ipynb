## 1. Introduction to Generative AI and the Rise of LLMs

- Generative AI has taken off since 2022.
- Key players: OpenAI (GPT-4), Anthropic (Claude 3), Meta (LLaMA), Google (Gemini), Amazon-backed initiatives.
- Goal: Understand differences between AI, ML, DL, and Generative AI.

## 2. What is Artificial Intelligence (AI)?

- AI = Applications that perform tasks without human intervention.
- Examples:
  - Netflix: Recommendation system
  - Self-driving cars
- AI Engineers integrate models into scalable software like web/mobile apps.

## 3. What is Machine Learning (ML)?

- Subset of AI.
- Provides statistical tools for:
  - Data analysis
  - Prediction & forecasting
  - Visualization
- Helps understand data and make it actionable.

## 4. What is Deep Learning (DL)?

- Subset of ML.
- Uses neural networks to mimic the human brain.
- Became popular due to GPUs, open-source tools, etc.
- Core techniques:
  - CNN: Computer vision
  - RNN/LSTM/GRU: Text/time series
  - Transformers & BERT: Foundation for Generative AI

## 5. What is Generative AI?

- Subset of Deep Learning.
- Focus: Generate new content (text, images, videos).
- Types of models:
  - Discriminative (e.g., classifiers)
  - Generative (e.g., text/image creators)
- Generative models are trained on massive datasets to generate new data.

## 6. Discriminative vs Generative Models

| Type           | Purpose                | Training Data     |
|----------------|------------------------|--------------------|
| Discriminative | Classification, etc.   | Labeled datasets   |
| Generative     | Content creation       | Huge unlabeled data|

## 7. Foundation Models (LLMs & LIMs)

- Examples:
  - OpenAI: GPT-4, GPT-5 (upcoming)
  - Anthropic: Claude 3
  - Meta: LLaMA 2/3
  - Google: Gemini, Gemma
- These are massive pre-trained models used across domains.
- Known as: Foundation models or pre-trained models.

## 8. Fine-Tuning Foundation Models

- Use your own data to adapt models to your needs.
- Techniques include:
  - LoRA (Low-Rank Adaptation)
- Enables domain-specific use without retraining from scratch.

## 9. LangChain and LLM (Large Language Model) Applications

- LangChain = Framework to connect LLMs to data/tools.
- Enables:
  - Chatbots
  - Retrieval-Augmented Generation (RAG)
  - Business workflows with AI automation

## 10. Large Image Models (LIMs)

- Like LLMs, but for image/video generation.
- Companies like Stability AI lead this field.
- Use cases: Text-to-image, video synthesis, etc.

## Conclusion

- Generative AI is reshaping app development and content creation.
- The race is on to build the best foundation models.
